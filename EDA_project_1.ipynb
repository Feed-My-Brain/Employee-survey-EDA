```python
# necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset

df = pd.read_csv("/content/drive/MyDrive/employee_survey.csv")
```

```python
# what are the data is occured inside our dataset

print(df.dtypes)
```

```python
# find or check the duplicate values

duplicates = df.duplicated().sum() # sum() it makes the sum of the all values

print(f'Number of duplicates inside the dataset: {duplicates}')
```

```python
# gender distribution

sns.countplot(x='Gender' , data=df)
plt.title("gender distribution")
plt.show()
```

```python
sns.countplot(x='MaritalStatus' , data=df)
plt.title("MaritalStatus distribution")
plt.show()
```

```python
sns.histplot(df['Age'] , kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# TrainingHoursPerYear

sns.histplot(df['TrainingHoursPerYear'] , kde=True)
plt.title("Training Hours Per Year")
plt.show()
```

```python
# SleepHours

sns.boxplot(x=df['SleepHours'])
plt.title("Sleep Hours")
plt.show()
```

```python
# Workload

sns.boxplot(x=df['Workload'])
plt.title("Work load")
plt.show()
```

```python
# select the numerical data

num_df = df.select_dtypes(include=['float64', 'int64'])

# increase the size of plot
plt.figure(figsize=(15,10))

# heatmap

sns.heatmap(num_df.corr() , cmap='coolwarm' , annot=True)
plt.title("correlation of numerical data")
plt.show()
```
